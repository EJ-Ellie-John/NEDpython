```python
X=df.drop(columns=['placement'])
X
Y=df['placement']
Y
```

```python
# train train split

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=90)

X_train
```

```python
X_test
```

To remove bias from data

As data with higher numerical value is given important under normal circumstances
min max scaler : use it when you want to normalize
standard scaler : use it when you want to standardize

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

Restart Kernel after scaling as the data is updated, you can reverse it

**"Can we"**

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_scaled,Y_train)
```

```python
Y_pred = clf.predict(X_test_scaled)
Y_pred
```

```python
pd.DataFrame({'Y_test':Y_test,'Y_predict':Y_pred})
```
