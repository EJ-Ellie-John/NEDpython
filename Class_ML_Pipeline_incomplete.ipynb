```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']

X
y
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

X_train.shape
```

```python
# imputation transformer i.e. fillna

trf1 = ColumnTransformer([
    ('impute_age', SimpleImputer(), ['Age']), #column index 2
    ('impute_embarked', SimpleImputer(strategy='most_frequent'), ['Embarked']) # mode is most frequent
], remainder='passthrough', verbose_feature_names_out=False)
```

```python
trf1
```

```python
# trf2 preprocess numerical columns and categorical columns separately

trf2 = ColumnTransformer([
    ('scale', MinMaxScaler(), slice(0,10)) # column index 0 to 9
], remainder='passthrough')

# preprocess categorical columns gender and embarked

trf3 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse_output=False, drop='first'), ['Sex', 'Embarked']) # column index is 1 and 6
], remainder='passthrough', verbose_feature_names_out=False)
```

```python
# scaling apply standard scaler

trf4 = ColumnTransformer([
    ('scale', StandardScaler(), slice(0,10)) # column index 0 to 9
], remainder='passthrough')
```

```python
trf4
```

```python
# Feature Selection

#define selectkbest
from sklearn.feature_selection import SelectKBest, chi2, f_classif
trf5 = SelectKBest(score_func=f_classif, k='all')
```

```python
trf5
```

```python
# model selection transformer

trf6 = DecisionTreeClassifier()
```

```python
trf6
```

```python
# Create pipeline

PL = Pipeline(
    [
     ('trf1', trf1),
     ('trf2', trf2),
     ('trf3', trf3),
     ('trf4', trf4),
     ('trf5', trf5),
     ('trf6', trf6)
    ]
)
```

```python
PL
```

Now we have wrapped it into a **pipeline**

```python
#Train the model
PL.fit(X_train, y_train)
```
