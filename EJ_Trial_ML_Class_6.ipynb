```python
X=df.drop(columns=['placement'])
X
Y=df['placement']
Y
```

```python
# train train split

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=90)

X_train
```

```python
X_test
```

To remove bias from data

As data with higher numerical value is given important under normal circumstances
min max scaler : use it when you want to normalize
standard scaler : use it when you want to standardize

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

Restart Kernel after scaling as the data is updated, you can reverse it

**"Can we"**

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train_scaled,Y_train)
```

```python
Y_pred = clf.predict(X_test_scaled)
Y_pred
```

```python
pd.DataFrame({'Y_test':Y_test,'Y_predict':Y_pred})
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
```

```python
print("Accuracy Score:", accuracy_score(Y_test,Y_pred))
print("Recall Score:", recall_score(Y_test,Y_pred))
print("Precision Score:", precision_score(Y_test,Y_pred))
print("F1 Score:", f1_score(Y_test,Y_pred))
print("Confusion Matrix:", confusion_matrix(Y_test,Y_pred))
#print("Classification Report:", classification_report(Y_test,Y_pred))
```

```python
sns.heatmap(confusion_matrix(Y_test,Y_pred), annot=True)
plt.show()
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
from mlxtend.plotting import plot_confusion_matrix
fig, ax = plot_confusion_matrix(conf_mat=confusion_matrix(Y_test,Y_pred))
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X=X, Y=Y_pred, clf=clf, legend=2)
plt.show()
```
